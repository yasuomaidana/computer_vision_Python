```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

## Building a neural network with secuence

```python
model = Sequential([
    #Dense(#of neurons that end on the layer, input_shape=(size of the shape),activation='activation method')
    Dense(5, input_shape=(3,),activation='relu'),
    #Dense(#of neurons on layer,#we removed imput shape,activation='activation method')
    Dense(2,activation='softmax'),
])
```

## Understanding activations

```python
model=Sequential() #Creating an empty neuron network
model.add(Dense(5, input_shape=(3,)))#Creates a layer withouth activation
model.add(Activation('relu'))#adds an activation method to the layer
```
