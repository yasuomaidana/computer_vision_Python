```python
#Mini tutorial
xx = "guru99,education is fun"
r1 = re.findall(r"^\w+",xx)
print(r1)
r1 = re.findall(r"\w+",xx)
print(r1)
#r1 = re.findall(r"\w",xx)#Encuentra las condiciones
#print(r1)
r1 = re.findall(r"^\w",xx)
print(r1)
r1 = re.split(r"\s",xx)#divide cuando se cumple el argumento \s espacio
print(r1)
r1 = re.split(r"\d",xx)#\d numero
print(r1)
xx=["guru not working","guru casa","guru guardia","guru ejemplo","casa del guru"]
for i in xx:
    a=re.match("(g\w+)\W(n\w+)",i) #si encuentra una coincidencia se prende la bandera
    b=re.match("(c\w+)\W(\w)",i) # a.groups devuelve los elementos que cumplieron la condicion
    if a:
        print('a')
        print(a.groups())
    if b:
        print('b')
        print(b.groups())
    if re.search("guru guardia",i):
        print(i)
        print("Encontro el patron")
        
abc = 'guru99@google.com, careerguru99@hotmail.com, users@yahoomail.com ,9Casa de de'
emails = re.findall(r'[\w\.-]+@[\w\.-]+', abc)

for email in emails:
    print(email)

```

```python
prueba = re.findall(r'9[\w.-]+\s', abc)
for email in prueba:
    print(email)
prueba = re.findall(r'y[\w.-]+\.', abc)
for email in prueba:
    print(email)
```
