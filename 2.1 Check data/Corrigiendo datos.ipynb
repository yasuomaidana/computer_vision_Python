# Corrigiendo con scalers


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = MinMaxScaler()
```

```python
print(scaler.fit(data))
```

```python
print(scaler.data_max_)#Muestra el dato mas grande
print(scaler.data_min_)#Muestra el dato mas pequeno
```

```python
print('Before')
print(data) #En forma de lista de coordenadas
print('After')
print(scaler.transform(data))
```

```python
print(scaler.transform([[2, 2]]))
```
