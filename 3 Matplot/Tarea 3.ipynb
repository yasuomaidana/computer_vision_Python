## Cuarto video
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.barh.html#matplotlib.pyplot.barh
Libreria utilizada

```python
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos = np.arange(len(company)) #Usado como guia para los nombres
#Se tiene que hacer esto porque bar() no acepta arreglos de string, solo valores numericos

plt.bar(xpos,revenue, label="Revenue")
#Hace grafica de barras con los valores de revenue, siguiendo el orden de xpos
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
#Hace grafico de barras asignando un nombre a cada columna
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bn)') #Anade el nombre al eje y
plt.title('US Teck Stock') #Anade titulo
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bn)') 
plt.title('US Teck Stock')

plt.bar(xpos,revenue,label="Revenue") #Imprime una grafica de barras
plt.bar(xpos,profit,label="Profit") #Superpone la grafica de barras
plt.legend()#Muestra los labels
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bn)') 
plt.title('US Teck Stock') 

plt.bar(xpos-.2,revenue,label="Revenue")#Mueve las barras hacia la izquierda
plt.bar(xpos+.2,profit,label="Profit")#Mueve las barras hacia la derecha
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bn)') 
plt.title('US Teck Stock') 

plt.bar(xpos-.2,revenue,width=.4,label="Revenue")#Modifica el grosor de las barras 
plt.bar(xpos+.2,profit,width=.4,label="Profit")#Modifica el grosor de las barras 
plt.legend()
```

```python
plt.barh(xpos-.2,revenue,height=0.4,label="Revenue")#Hace grafica de barras en forma horizontal, height modifica el grosor
plt.barh(xpos+.2,profit,height=0.4,label="Profit")#Hace grafica de barras en forma horizontal 
```

```python
plt.yticks(xpos,company)#Coordina los nombres con la posicion en el eje y
plt.xlabel('revenue(bn)')#Le da nombre al eje x
plt.title('US Teck Stock')

plt.barh(xpos-.2,revenue,height=0.4,label="Revenue")
plt.barh(xpos+.2,profit,height=0.4,label="Profit")
plt.legend()
```

## Quinto video
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) #Hace histograma
```

```python
plt.hist(blood_sugar,bins=5) # bins controla el numero de segmentos
```

```python
plt.hist(blood_sugar,bins=5,width=13) #width controla el grosor de las barras
```

```python
plt.hist(blood_sugar,bins=5,rwidth=.95) #rwidth controla la distancia entre barras
```

```python
plt.hist(blood_sugar,bins=[77,100,120,150],rwidth=.95) #Hace los segmentos mediante un arreglo que marca los intervalos
```

```python
plt.hist(blood_sugar,bins=[77,100,120,150],rwidth=.95,color='purple')#controla el color
```

```python
plt.hist(blood_sugar,bins=[77,100,120,150],rwidth=.95,color='purple',histtype='step')
#Convierte el histograma a un tipo step
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95) #Grafica 2 datos mediante histograma
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['blue','purple'])
#Controla el color mediante color
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
#Anade lables
plt.legend()#Muestra los labels
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
#Asigna nombres a los ejes y da titulo

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.ylabel("Sugar Level")
plt.xlabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['blue','purple'],
        label=['men','women'],orientation='horizontal') #Modifica la orientacion de la grafica
plt.legend()
plt.show()#Muestra la grafica
```

## Sexto video
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels) #Grafica una grafica de pastel
plt.show() 
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5) #Modifica el radio de la grafica de pastel
plt.show() 
```

```python
plt.axis('equal') #Hace que los ejes midan lo mismo
plt.pie(exp_vals,labels=exp_labels,radius=1.5) 
plt.show() 
```

```python
plt.axis('equal') 
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%')#Muestra el porcentaje de cada seccion 
plt.show() 
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True)#Muestra sombra de la grafica
plt.show() 
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,.2,0,0,0])
#Hace un explosionado a la distancia que se le pone en el vector 
plt.show() 
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,.2,0,0,0],startangle=90)
#Modifica el angulo de inicio
plt.show() 
```

## Video 7
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.savefig.html#matplotlib.pyplot.savefig
Libreria utilizada

```python
from PIL import Image #Nos permitira mostrar las imagenes creadas
import webbrowser
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',shadow=True,explode=[0,.2,0,0,0],startangle=90)
plt.savefig("piechart.png")#Guarda la grafica con el nombre que se le asigna
Image.open(r"piechart.png").show()#Muestra la grafica guardada

plt.savefig("piechartC.png",bbox_inches="tight")#Se asegura que entre en la figura
Image.open(r"piechartC.png").show()

plt.savefig("piechartE.png",bbox_inches="tight",pad_inches=2)#Anade un espacio en blanco
Image.open(r"piechartE.png").show()

plt.savefig("piechartT.png",bbox_inches="tight",pad_inches=2,transparent=True)#Hace que el fondo sea transparente
Image.open(r"piechartT.png").show()

plt.savefig("piechart.pdf",bbox_inches="tight",pad_inches=2,transparent=True)#Guarda el archivo como pdf
open("piechart.pdf") #Abre el archivo 
webbrowser.open_new(r'piechart.pdf')#Abre el archivo pdf en el explorador
```
