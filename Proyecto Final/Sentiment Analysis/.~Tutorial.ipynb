```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize #chops a sentence or paragraph
from nltk.corpus import stopwords #get stop words, common words that doesnt say anything
from nltk.stem import PorterStemmer #convert similar words (conjugated one) in only one variable
from nltk.corpus import state_union #
from nltk.tokenize import PunktSentenceTokenizer #trained algorithm
from nltk import pos_tag #assign tag (type)
from nltk import RegexpParser #regular expresion chopper/parser
from nltk import ne_chunk#assign tags (name, organization etc..)
#https://pythonprogramming.net/part-of-speech-tagging-nltk-tutorial/
from nltk.stem import WordNetLemmatizer #convert similar words into a class
from nltk.corpus import wordnet
from nltk.corpus import movie_reviews
from nltk.classify.scikitlearn import SklearnClassifier
from nltk.classify import ClassifierI #permite herencias

import random
import pickle
from statistics import mode


from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC

def find_features(document):
    words=set(document)
    features={}
    for w in word_features:
        features[w]=(w in words)
    return features

class VoteClassifier(ClassifierI):
    def __init__(self, *classifiers):
        self._classifiers = classifiers

    def classify(self, features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)
        return mode(votes)

    def confidence(self, features):
        votes = []
        for c in self._classifiers:
            v = c.classify(features)
            votes.append(v)

        choice_votes = votes.count(mode(votes))
        conf = choice_votes / len(votes)
        return conf
```

# Preparando info

```python
short_pos = open("positive.txt","r").read()
short_neg = open("negative.txt","r").read()

# Inicializando
all_words = []
documents = []

#J adjetivo R adverbio V verbo 
#allowed_word_types = ["J","R","V"]
allowed_word_types = ["J"]


###################################
###Filtrado de palabras
for p in short_pos.split('\n'): #parte por nueva linea
    documents.append( (p, "pos") ) #anade a la lista de documentos
    words = word_tokenize(p)#parte por palabras
    pos = nltk.pos_tag(words)#asigna tags a las pabras
    for w in pos:#para cada palabra tageada
        if w[1][0] in allowed_word_types:#verifica que se encuentre en el atributo deseado
            all_words.append(w[0].lower())

for p in short_neg.split('\n'):
    documents.append( (p, "neg") )
    words = word_tokenize(p)
    pos = nltk.pos_tag(words)
    for w in pos:
        if w[1][0] in allowed_word_types:
            all_words.append(w[0].lower())

#Guarda documento
save_documents = open("documents.pickle","wb")
pickle.dump(documents, save_documents)
save_documents.close()

all_words = nltk.FreqDist(all_words)#convierte en lista 
word_features = list(all_words.keys())[:5000]#selecciona a los primeros 5k

save_word_features = open("word_features5k.pickle","wb")
pickle.dump(word_features, save_word_features)
save_word_features.close()

#Arma los sets
featuresets = [(find_features(rev), category) for (rev, category) in documents]
#Revuelve los sets
random.shuffle(featuresets)
print(len(featuresets))

#Crea los sets de entrenamiento y prueba
testing_set = featuresets[10000:]
training_set = featuresets[:10000]

save_features = open("featuresets.pickle","wb")
pickle.dump(word_features, save_features)
save_word_features.close()
```
