## Using a web link to get information

```python
import requests
URL= 'https://www.monster.com/jobs/search/?q=Software-Developer&where=Australia'
page = requests.get(URL) #Obtiene la respuesta de la pagina
```

## Parse using BeautifulSoup

```python
from bs4 import BeautifulSoup as BS
```

```python
soup= BS(page.content,'html.parser')#la libreria parsea la informacion para que sea mas facil de utilizar
soup= BS(page.text,'html.parser')#hace lo mismo
print(soup.prettify())#.prettify prints in a nice way
```

## Find elements by ID

```python
findID= soup.find(id='ResultsContainer')#Busca la informacion con id='Whatever'
print(findID.prettify())
```

```python
print(type(findID))
```

## Find elements by HTML Class Name

```python
elements=findID.find_all('section',class_='card-content')
elements
```

```python
elements[1]
```

## Extract text from html elements

```python
for element in elements:
    title=element.find('h2',class_='title')
    company=element.find('div',class_='company')
    location=element.find('div',class_='location')
    if None in (title, company, location): #skips if it doesnt have any of them
        continue
    print(title.text.strip())#strip removes all the charactares that we don't want
    print(company.text.strip())
    print(location.text.strip())
    print()
```

```python
title=elements[0].find('h2',class_='title')
company=elements[0].find('div',class_='company')
location=elements[0].find('div',class_='location')
print(title.prettify())
print(company.prettify())
print(location.prettify())
```

## Find elements by class name and text content

```python
pythonJobs = findID.find_all('h2', string='Python Developer')
print(pythonJobs)
pythonJobs = findID.find_all('h2',string=lambda text: 'python' in text.lower())
print(pythonJobs)
print(len(pythonJobs))
```

```python
string=lambda text: 'python' in text.lower()
print(string('Casa'))
print(string('pyThon'))
```

## Extract attributes from HTML elements

```python
for jobs in pythonJobs:
    link=jobs.find('a')['href']
    print(jobs.text.strip())
    print(f"Apply here: {link}\n")
```
